```python
import pandas as pd
import requests
from io import StringIO
import random
import time
import re
import numpy as np
import urllib.parse as urllib
import urllib.request as urllib2
import json
import time
from bs4 import BeautifulSoup
```

```python
s = []
restaurant = []
restaurant_add = []
for i in range(1,36):
    url = 'https://www.ipeen.com.tw/search/all/000/1-0-0-0/?p={}&adkw=%E6%9E%97%E5%8F%A3'.format(i)
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    r = requests.get(url, headers)
    r.encoding = 'utf-8'
    soup = BeautifulSoup(r.text,'html.parser')    
    for i in soup.findAll('a','a37 ga_tracking'):
        if i['data-label'] == '店名':
            restaurant.append(i.text)
            print(i.text)
    for i in soup.findAll('span','address'):
        restaurant_add.append(i.text.split('\n')[0])
```
